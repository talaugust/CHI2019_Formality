```python
def descr_df(df):
    ### Descriptive Stats on Demographics and Motivations ###
    print(len(df))
    print(df['formality_level'].unique())
    print('Turker?', df['Turker'].unique())
    print('Descriptive stats - Demographics')
    print('------------------------------------------------' )
    print('Age')
    print('------------------------------------------------' )
    print('Mean age:', df['age'].mean())
    print('Std age:', df['age'].std())
    print('------------------------------------------------' )
    print('Gender')
    print('------------------------------------------------' )
    print('Gender breakdown:')
    print(df['gender'].value_counts())
    print('% Female', df['gender'].value_counts()['female']/len(df['gender']))
    print('------------------------------------------------' )
    print('Country')
    print('------------------------------------------------' )
    print('Country counts:')
#     print(df['country0'].value_counts())
    print(len(df['country0'].value_counts()))
    print('% English:', len(df[df['english'] == 1])/len(df))
    print('------------------------------------------------' )
    print('Education')
    print('------------------------------------------------' )
    print('Education levels:')
#     print(df['education'].value_counts())
    print(df['education'].mean())
    print(df['education'].std())
    print('------------------------------------------------' )

```

```python
### Descriptive stats total
df_cleaned = df_dem_formality_values.dropna(subset=['gender', 'age', 'education'])
descr_df(df_cleaned)
len(df_cleaned[df_cleaned['english'] == 1])/len(df_cleaned)
```

```python
### Descriptive stats for formal vs. informal
formality_groups = df_cleaned.groupby('formality_level')
formality_groups.apply(descr_df)

```

```python
### Descriptive stats for turkers
turker_groups = df_cleaned.groupby('Turker')
turker_groups.apply(descr_df)

```

```python
def run_formal_vs_informal_descr_stats(df):
    print('Formal')
    print('------------------------------------------------' )
    df_formal = df[df['formality_level'] == 'formal']
    descr_df(df_formal)
    print('------------------------------------------------' )
    print('Informal')
    print('------------------------------------------------' )
    df_informal = df[df['formality_level'] == 'informal']
    descr_df(df_informal)
   
```

```python
turker_groups.apply(run_formal_vs_informal_descr_stats)
```
